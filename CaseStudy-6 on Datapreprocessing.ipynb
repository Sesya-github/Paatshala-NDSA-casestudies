# 4.) Fill in all the missing values present in all the columns in the dataset

```python
data.isna().sum()
```

The features Age, Cabin, Embarked have missing values.

```python
print('Percentage of missing values in Age Column: ',100-(data.Age.count()*100/len(data)))
print('Percentage of missing values in Cabin Column: ',100-(data.Cabin.count()*100/len(data)))
print('Percentage of missing values in Embarked Column: ',100-(data.Embarked.count()*100/len(data)))
```

Age is a numerical variable, let us see its distriution

```python
sns.distplot(data.Age)
```

The distribution does not seem to be highly skewed. Futher we can classify the passengers according to their passenger class and assign mean age value of each class to the missing values

```python
data.groupby('Pclass')['Age'].mean()
```

```python
#function to assign mean age of each passenger class to those missing age value
def age_assign(sliced):
    pc= sliced[0]
    age=sliced[1]
    if pd.isnull(age):
        if pc==1:
            return(38)
        elif pc==2:
            return(30)
        else:
            return(25)
    else:
        return age
```

```python
data['Age']=data[['Pclass','Age']].apply(age_assign,axis=1)
data
```

Cabin feature has nearly 77% missing values, and there is no problem if we drop this feature from the data.

```python
#droping the feature Cabin
data.drop('Cabin',axis=1,inplace=True)
```

Embarked feature contain only 2(0.22%) missing values, and the dataset is considerably large so we can drop those observations which has null value in Embarked

```python
data.dropna(subset=['Embarked'],inplace=True)
# We shall not reset the index after dropping the row, as the data is indexed on passenger id
data
```

```python
data.isna().sum()
```

Now there is no missing values present in the dataset

# 5.) Check and handle outliers in at least 3 columns in the dataset

```python
sns.boxplot(data=data)
```

There are outliers in features Age, SibSp, Parch, Fare. Let us handle outliers in Age, SibSp and Fare

```python
sns.boxplot(data.Age)
```

Age of Passengers range between 0 and 80. This doesnt seem odd. So keep the values of age as it is.

```python
sns.boxplot(data.SibSp)
```

 SibSp ranges between 0 to 8. Since the number of Siblings+Spouse values seems reasonably natural, this too can be let as it is.

```python
sns.boxplot(data.Fare)
```

In the Fare column, only one value looks extremely as high as 500. This seems odd.So the extreme shall be replaced by the median value as the fare seems kind of left skewed.

```python
Q1 = np.percentile(data['Fare'],25)
median = np.percentile(data['Fare'],50)
Q3 = np.percentile(data['Fare'],75)
IQR = Q3-Q1
outlier_index = data[(data['Fare']>(Q3+1.5*IQR)) | (data['Fare']<(Q1-1.5*IQR))].index
data.loc[list(outlier_index),'Fare']=median
sns.boxplot(x=data.Fare)
```
